### 4. Часто на практике нулевая гипотеза отклоняется, и различия считаются статистически достоверными, если p < 0,05. Однако часто в статистике используется более жесткий критерий достоверности различий, например, при условии, что p < 0,01. Значение p-уровня значимости, которое выбирается, в качестве порога обозначается буквой α (альфа). Например, если исследователь решил, что α = 0,05, то и нулевая гипотеза будет отклоняться при условии, что p < 0,05. 

### Если в определенной ситуации весьма рискованно отклонить нулевую гипотезу, когда она на самом деле верна, то лучше использовать показатель α равный

*Если в определенной ситуации весьма рискованно отклонить нулевую гипотезу, когда она на самом деле верна, то лучше использовать показатель α равный **0,001**.*

*Уровень значимости (α) представляет собой пороговое значение, используемое для определения статистической значимости результатов теста. Когда мы проводим статистические тесты, у нас всегда есть вероятность совершить ошибку первого рода (ошибка отклонения нулевой гипотезы, когда она на самом деле верна). Уровень значимости (α) указывает на вероятность совершения этой ошибки.*

*Обычно используемый уровень значимости в статистике - α = 0,05, что означает, что мы готовы принять риск ошибки первого рода в 5% случаев. Однако, в некоторых ситуациях, особенно в более важных и ответственных исследованиях, риски ошибок могут быть более значимыми.*

*Если существует высокий риск принять альтернативную гипотезу (отклонить нулевую гипотезу), когда она на самом деле верна (то есть, существует реальное отличие отвергаемой гипотезы), более предпочтительным может быть установление более низкого значения α, например, α = 0,001. Это означает, что мы готовы принять риск ошибки первого рода только в 0,1% случаев, что снижает вероятность ошибочного отклонения нулевой гипотезы в пользу более консервативной оценки результатов исследования.*
